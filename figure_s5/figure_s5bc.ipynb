```python
### Combine UHGV and UHVDB metadata to identify uncertain and confident viruses in UHVDB HQ+ dereplicated genomes

import polars as pl
# identify all fasta IDs
# !zgrep "^>" ../uhvdb_creation/uhvdb.vclust_derep_reps.fna.gz > uhvdb_unique_ids.txt
uhvdb_unique_ids = set(
    pl.read_csv('uhvdb_unique_ids.txt', has_header=False, new_columns=['seq_header'])
        .with_columns([
            pl.col('seq_header').str.replace('^>', '').str.split(' ').list[0].alias('seq_name')
        ])['seq_name']
)

# download UHGV metadata
# !wget https://portal.nersc.gov/cfs/m342/UHGV/metadata/uhgv_metadata.tsv
uhgv_hq_plus_derep = (
    pl.read_csv("../figure_s1/uhgv_metadata.tsv", separator="\t",
        columns=['uhgv_genome', 'viral_confidence', 'genomad_virus_hallmarks', 'genomad_plasmid_hallmarks', 'genomad_virus_score'], ignore_errors=True)
        .filter(pl.col('uhgv_genome').is_in(uhvdb_unique_ids))
        .rename({'uhgv_genome': 'seq_name'})
)

# filter mine report to only those in unique seq ids
mine_report_uhvdb_derep_hq = (
    pl.read_csv('../figure_1/viruses.csvtk_concat.tsv', separator='\t', columns=['seq_name', 'uhvdb_virus_classification', 'n_hallmarks', 'plasmid_hallmarks', 'virus_score'])
        .filter(pl.col('seq_name').is_in(uhvdb_unique_ids))
        .unique('seq_name')
        .rename({'uhvdb_virus_classification': 'viral_confidence', 'n_hallmarks': 'genomad_virus_hallmarks', 'plasmid_hallmarks': 'genomad_plasmid_hallmarks', 'virus_score': 'genomad_virus_score'})
        .with_columns([
            pl.col('viral_confidence').str.replace('confident', 'Confident')
        ])
        .with_columns([
            pl.col('viral_confidence').str.replace('uncertain', 'Uncertain'),
        ])
)
print("Number of new dereplicated HQ virus sequences in UHVDB:", mine_report_uhvdb_derep_hq.height)
# combine UHVDB and UHGV metadata
uhgv_uhvdb_hq_derep = pl.concat([
    uhgv_hq_plus_derep[['seq_name', 'viral_confidence', 'genomad_virus_hallmarks', 'genomad_plasmid_hallmarks', 'genomad_virus_score']],
    mine_report_uhvdb_derep_hq[['seq_name', 'viral_confidence', 'genomad_virus_hallmarks', 'genomad_plasmid_hallmarks', 'genomad_virus_score']]
    ]).unique('seq_name')

# count uncertain vs confident genomes
print(uhgv_uhvdb_hq_derep.group_by('viral_confidence').len())
```

```python
### Plot number of MMSEeqs2 virus hallmarks in UHVDB HQ+ dereplicated genomes
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.boxplot(data=uhgv_uhvdb_hq_derep, x="viral_confidence", y="genomad_virus_hallmarks", showmeans=True)
plt.title("geNomad virus hallmarks in HQ+ genomes")
plt.ylabel("Number of geNomad virus hallmarks")
plt.xlabel("Viral confidence category")
plt.axhline(3, color='r', linestyle='dashed')

maxes = uhgv_uhvdb_hq_derep.group_by(['viral_confidence']).agg([pl.col('genomad_virus_hallmarks').max()]).sort('viral_confidence')['genomad_virus_hallmarks']
nobs = uhgv_uhvdb_hq_derep['viral_confidence'].value_counts().sort('viral_confidence')['count']

# Add it to the plot
pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
    ax.text(pos[tick],
            maxes[tick] + 5,
            "n = " + str(nobs[tick]),
            horizontalalignment='center',
            size='small',
            color='black',
            weight='semibold')

plt.ylim(0, 95)
```

```python
### Plot number of MMSEeqs2 plasmid hallmarks in UHVDB HQ+ dereplicated genomes
ax = sns.boxplot(data=uhgv_uhvdb_hq_derep, x="viral_confidence", y="genomad_plasmid_hallmarks", showmeans=True)
plt.title("geNomad plasmid hallmarks in HQ+ genomes")
plt.ylabel("Number of geNomad plasmid hallmarks")
plt.xlabel("Viral confidence category")
plt.axhline(1, color='r', linestyle='dashed')

maxes = uhgv_uhvdb_hq_derep.group_by(['viral_confidence']).agg([pl.col('genomad_plasmid_hallmarks').max()]).sort('viral_confidence')['genomad_plasmid_hallmarks']
nobs = uhgv_uhvdb_hq_derep['viral_confidence'].value_counts().sort('viral_confidence')['count']

# Add it to the plot
pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
    ax.text(pos[tick],
            maxes[tick] + 3,
            "n = " + str(nobs[tick]),
            horizontalalignment='center',
            size='small',
            color='black',
            weight='semibold')

plt.ylim(0, 55)
```

```python
### Plot geNomad virus score in UHVDB HQ+ dereplicated genomes
ax = sns.boxplot(data=uhgv_uhvdb_hq_derep, x="viral_confidence", y="genomad_virus_score", showmeans=True)
plt.title("geNomad virus Score in HQ+ genomes")
plt.ylabel("geNomad virus score")
plt.xlabel("Viral confidence category")
plt.axhline(0.95, color='r', linestyle='dashed')

maxes = uhgv_uhvdb_hq_derep.group_by(['viral_confidence']).agg([pl.col('genomad_virus_score').max()]).sort('viral_confidence')['genomad_virus_score']
nobs = uhgv_uhvdb_hq_derep['viral_confidence'].value_counts().sort('viral_confidence')['count']

# Add it to the plot
pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
    ax.text(pos[tick],
            maxes[tick] + 0.01,
            "n = " + str(nobs[tick]),
            horizontalalignment='center',
            size='small',
            color='black',
            weight='semibold')
```

```python
### Count number of virus/plasmid hallmarks per genome from hmm results
import polars as pl
import glob

# count hmm hallmark hits per genome
genomad_virus_hallmarks = set(
    pl.read_csv('genomad_metadata_v1.9.tsv.gz', separator='\t', ignore_errors=True)
    .filter(
        (pl.col('VIRUS_HALLMARK') == 1)
    )['MARKER']
)

genomad_plasmid_hallmarks = set(
    pl.read_csv('genomad_metadata_v1.9.tsv.gz', separator='\t', ignore_errors=True)
    .filter(
        (pl.col('PLASMID_HALLMARK') == 1)
    )['MARKER']
)

results = []
for file in glob.glob('hmm_results/*.tbl'):
    with open(file, 'r') as tbl:
        for line in tbl:
            if '#' in line[0]:
                continue
            strip_split = line.strip().split()
            protein = strip_split[0]
            genome = protein.rsplit('_', 1)[0]
            target = strip_split[2]
            results.append({'genome': genome, 'protein': protein, 'hallmark': target, 'evalue': float(strip_split[4])})
    tbl.close()
```

```python
### summarize hallmarks per uncertain genome
uncertain_hmm_hallmarks = (
    pl.DataFrame(results)
        .with_columns([
            pl.when(pl.col('hallmark').is_in(genomad_virus_hallmarks)).then(1).otherwise(0).alias('virus_hallmarks'),
            pl.when(pl.col('hallmark').is_in(genomad_plasmid_hallmarks)).then(1).otherwise(0).alias('plasmid_hallmarks'),
        ])
        .sort('evalue', descending=False)
        .group_by('protein')
        .first()
        .group_by(['genome'])
        .agg([pl.col('virus_hallmarks').sum().alias('virus_hallmarks'), pl.col('plasmid_hallmarks').sum().alias('plasmid_hallmarks')])
        .join(uhgv_uhvdb_hq_derep, left_on='genome', right_on='seq_name')
)

ax = sns.boxplot(data=uncertain_hmm_hallmarks, x="plasmid_hallmarks", y='virus_hallmarks', showmeans=True)
plt.title("HMM hallmarks in uncertain genomes")
plt.ylabel("Virus hallmarks")
plt.xlabel("Plasmid hallmarks")
plt.axhline(3, color='r', linestyle='dashed')
plt.axvline(0.5, color='r', linestyle='dashed')

maxes = uncertain_hmm_hallmarks.group_by(['plasmid_hallmarks']).agg([pl.col('virus_hallmarks').max()]).sort('plasmid_hallmarks')['virus_hallmarks']
nobs = uncertain_hmm_hallmarks['plasmid_hallmarks'].value_counts().sort('plasmid_hallmarks')['count']

# Add it to the plot
pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
    if pos[tick] == 2:
        maxes[tick] = 55
    ax.text(pos[tick],
            maxes[tick] + 5,
            "n = " + str(nobs[tick]),
            horizontalalignment='center',
            size='small',
            color='black',
            weight='semibold')
plt.ylim(0, 75)
plt.xlim(-1, 12)
```

```python
### write out genomes with >= 3 virus hallmarks and 0 plasmid hallmarks (would add 1 point to score)
uncertain2confident = (
    uncertain_hmm_hallmarks
        .filter(
            (pl.col('virus_hallmarks') >= 3) &
            (pl.col('plasmid_hallmarks') == 0)
        )
)

print("Number of sequences that would move from Uncertain to Confident:", uncertain2confident.shape[0])

uncertain2confident[['genome']].write_csv('uhvdb_hq_derep_uncertain2confident.txt', include_header=False)
```
